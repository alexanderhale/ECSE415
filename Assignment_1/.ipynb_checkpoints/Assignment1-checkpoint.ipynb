## Question 1 - Thresholding

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [14, 7]
```

```python
# load image and convert to grayscale
img_num_gray = cv2.cvtColor(cv2.imread("./numbers.jpg"), cv2.COLOR_BGR2GRAY)
plt.imshow(img_num_gray, cmap="gray")
plt.title("Original Greyscale Image"), plt.xticks([]), plt.yticks([])
plt.show()

# threshold values
thresh1 = 55
thresh2 = 90
thresh3 = 150

# create output images
ret, img_thresh1 = cv2.threshold(img_num_gray, thresh1, 255, cv2.THRESH_BINARY)
ret, img_thresh2 = cv2.threshold(img_num_gray, thresh2, 255, cv2.THRESH_BINARY)
ret, img_thresh3 = cv2.threshold(img_num_gray, thresh3, 255, cv2.THRESH_BINARY)

# display images
plt.subplot(131), plt.imshow(img_thresh1, cmap="gray")
plt.title("T = 55"), plt.xticks([]), plt.yticks([])
plt.subplot(132), plt.imshow(img_thresh2, cmap="gray")
plt.title("T = 90"), plt.xticks([]), plt.yticks([])
plt.subplot(133), plt.imshow(img_thresh3, cmap="gray")
plt.title("T = 150"), plt.xticks([]), plt.yticks([])
plt.show()
```

At a threshold of T = 55, all of the numbers are added to the foreground. Additionally, some of the lighter parts of the original image that are not numbers have intensities high enough to exceed the threshold, so a circle is added to the foreground. This circle obscures the number 549 because both the circle and the number are considered part of the foreground.

At T = 90, 5 of the 6 numbers are segmented. The number 956 in the original image has an intensity too low to be considered part of the foreground in the threshold image.

At T = 150, 3 of the 6 numbers are segmented. The numbers 956, 627, and 845 have intensity values lower than the threshold.

```python
# threshold value to only include 123 and 549 as part of the foreground
thresh4 = 205

# create thresholded image
ret, img_thresh4 = cv2.threshold(img_num_gray, thresh4, 255, cv2.THRESH_BINARY)

# display image
plt.imshow(img_thresh4, cmap="gray")
plt.title("T = " + str(thresh4)), plt.xticks([]), plt.yticks([])
plt.show()
```

At a threshold value of T = 205, we achieve the goal of segmenting only 123 and 549.

## Question 2 - Filtering
### 2.1 - Denoising

```python
# load image and convert to grayscale
img_cir_clean = cv2.cvtColor(cv2.imread("./Circles.png"), cv2.COLOR_BGR2GRAY)
img_cir_gray = cv2.cvtColor(cv2.imread("./Circles_gauss.png"), cv2.COLOR_BGR2GRAY)

# filter image with a 5x5 Gaussian filter
img_gauss = cv2.GaussianBlur(img_cir_gray,(5,5),0)

# filter image with a 5x5 box filter
img_box = cv2.blur(img_cir_gray, (5,5))

# display images
plt.subplot(121), plt.imshow(img_cir_clean, cmap="gray")
plt.title("Clean Image"), plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(img_cir_gray, cmap="gray")
plt.title("Noisy Image"), plt.xticks([]), plt.yticks([])
plt.show()
print("Unfiltered PSNR: " + str(cv2.PSNR(img_cir_clean, img_cir_gray)))

plt.subplot(121), plt.imshow(img_gauss, cmap="gray")
plt.title("5x5 Gaussian Filter"), plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(img_box, cmap="gray")
plt.title("5x5 Box Filter"), plt.xticks([]), plt.yticks([])
plt.show()
print("Gaussian Filtered PSNR: " + str(cv2.PSNR(img_cir_clean, img_gauss)))
print("Box Filtered PSNR: " + str(cv2.PSNR(img_cir_clean, img_box)))
```

The PSNR for the Gaussian filtered image is higher than the PSNR for the box-filtered image. Since the MSE term is in the denominator of the function for PSNR, a higher PSNR is a better result. Therefore, the Gaussian filter performs better than the box filter. This is logical because the error introduced to the image is Gaussian in nature.

```python
# filter image with 9x9 and 13x13 filters
img_gauss_9 = cv2.GaussianBlur(img_cir_gray,(9,9),0)
img_gauss_13 = cv2.GaussianBlur(img_cir_gray,(13,13),0)
img_box_9 = cv2.blur(img_cir_gray, (9,9))
img_box_13 = cv2.blur(img_cir_gray, (13,13))

# display images
plt.subplot(121), plt.imshow(img_gauss_9, cmap="gray")
plt.title("9x9 Gaussian Filter"), plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(img_gauss_13, cmap="gray")
plt.title("13x13 Gaussian Filter"), plt.xticks([]), plt.yticks([])
plt.show()
print("9x9 Gaussian Filtered PSNR: " + str(cv2.PSNR(img_cir_clean, img_gauss_9)))
print("13x13 Gaussian Filtered PSNR: " + str(cv2.PSNR(img_cir_clean, img_gauss_13)))

plt.subplot(121), plt.imshow(img_box_9, cmap="gray")
plt.title("9x9 Box Filter"), plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(img_box_13, cmap="gray")
plt.title("13x13 Box Filter"), plt.xticks([]), plt.yticks([])
plt.show()
print("9x9 Box Filtered PSNR: " + str(cv2.PSNR(img_cir_clean, img_box_9)))
print("13x13 Box Filtered PSNR: " + str(cv2.PSNR(img_cir_clean, img_box_13)))
```

For both filters, the PSNR dropped as the dimension of the filter increased, meaning that the higher dimension reduced performance. The Gaussian filter still outperformed the Gaussian filter (in terms of PSNR) for both the 9x9 and 13x13 trials.

Visually, the amount of noise diminishes as the dimensions of the filters improve, which would indicate that the higher-dimension filters are performing better. The box filter removes more noise than the Gaussian filter, at the cost of more blurring of the image. The filtering results in a decrease in visual image quality because the image is quite blurry.

```python
# bring in salt-and-pepper noise image
img_cir_sp = cv2.cvtColor(cv2.imread("./Circles_sp.png"), cv2.COLOR_BGR2GRAY)

# filter image with 5x5, 9x9, and 13x13 filters
img_gauss_5_sp = cv2.GaussianBlur(img_cir_sp,(5,5),0)
img_gauss_9_sp = cv2.GaussianBlur(img_cir_sp,(9,9),0)
img_gauss_13_sp = cv2.GaussianBlur(img_cir_sp,(13,13),0)
img_median_5_sp = cv2.medianBlur(img_cir_sp, 5)
img_median_9_sp = cv2.medianBlur(img_cir_sp, 9)
img_median_13_sp = cv2.medianBlur(img_cir_sp, 13)

# display images
plt.subplot(131), plt.imshow(img_cir_sp, cmap="gray")
plt.title("Unfiltered Salt-and-Pepper Noise"), plt.xticks([]), plt.yticks([])
plt.show()
print("Unfiltered PSNR: " + str(cv2.PSNR(img_cir_clean, img_cir_sp)))

plt.subplot(131), plt.imshow(img_gauss_5_sp, cmap="gray")
plt.title("5x5 Gaussian Filter"), plt.xticks([]), plt.yticks([])
plt.subplot(132), plt.imshow(img_gauss_9_sp, cmap="gray")
plt.title("9x9 Gaussian Filter"), plt.xticks([]), plt.yticks([])
plt.subplot(133), plt.imshow(img_gauss_13_sp, cmap="gray")
plt.title("13x13 Gaussian Filter"), plt.xticks([]), plt.yticks([])
plt.show()
print("5x5 Gaussian Filtered PSNR: " + str(cv2.PSNR(img_cir_clean, img_gauss_5_sp)))
print("9x9 Gaussian Filtered PSNR: " + str(cv2.PSNR(img_cir_clean, img_gauss_9_sp)))
print("13x13 Gaussian Filtered PSNR: " + str(cv2.PSNR(img_cir_clean, img_gauss_13_sp)))

plt.subplot(131), plt.imshow(img_median_5_sp, cmap="gray")
plt.title("5x5 Median Filter"), plt.xticks([]), plt.yticks([])
plt.subplot(132), plt.imshow(img_median_9_sp, cmap="gray")
plt.title("9x9 Median Filter"), plt.xticks([]), plt.yticks([])
plt.subplot(133), plt.imshow(img_median_13_sp, cmap="gray")
plt.title("13x13 Median Filter"), plt.xticks([]), plt.yticks([])
plt.show()
print("5x5 Median Filtered PSNR: " + str(cv2.PSNR(img_cir_clean, img_median_5_sp)))
print("9x9 Median Filtered PSNR: " + str(cv2.PSNR(img_cir_clean, img_median_9_sp)))
print("13x13 Median Filtered PSNR: " + str(cv2.PSNR(img_cir_clean, img_median_13_sp)))
```

All of the implemented filters make improvements (i.e. increase the PSNR) over the original salt-and-pepper noisy image. The PSNR for the median-filtered images is higher than that of the Gaussian-filtered images, meaning the median filter performs better when removing the salt-and-pepper noise. As the size of the Gaussian filter increases to 9x9 its performance improves, but further increasing the size to 13x13 decreases performance slightly. Increasing the size of the median filter beyond the 5x5 size to 9x9 and 13x13 decreases its performance.

The visual results correlate with the PSNR results. The 9x9 Gaussian filter provides the best balance out of the three Gaussian filters between noise reduction and blur. The 5x5 median filter provides the best balance out of all the images between noise reduction and blur. The median filter removes virtually all of the noise from the salt-and-pepper image, and the larger filter sizes serve only to blur the image further without removing any additional noise.

### 2.2 - Sharpening

```python
# bring in rice image
img_rice = cv2.cvtColor(cv2.imread("./rice.png"), cv2.COLOR_BGR2GRAY)

# create blurred images
rice_blur_box = cv2.blur(img_rice,(7,7))
rice_blur_gauss = cv2.GaussianBlur(img_rice,(7,7),0)

# output images
plt.subplot(121), plt.imshow(rice_blur_box, cmap="gray")
plt.title("Box Blur"), plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(rice_blur_gauss, cmap="gray")
plt.title("Gauss Blur"), plt.xticks([]), plt.yticks([])
plt.show()

# create detail images
rice_detail_box = cv2.subtract(img_rice, rice_blur_box)
rice_detail_gauss = cv2.subtract(img_rice, rice_blur_gauss)

# output images
plt.subplot(121), plt.imshow(rice_detail_box, cmap="gray")
plt.title("Box Detail"), plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(rice_detail_gauss, cmap="gray")
plt.title("Gauss Detail"), plt.xticks([]), plt.yticks([])
plt.show()

# create sharpened image
rice_sharp_box = cv2.add(img_rice, rice_detail_box)
rice_sharp_gauss = cv2.add(img_rice, rice_detail_gauss)

# output images
plt.subplot(121), plt.imshow(rice_sharp_box, cmap="gray")
plt.title("Box Sharpened"), plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(rice_sharp_gauss, cmap="gray")
plt.title("Gauss Sharpened"), plt.xticks([]), plt.yticks([])
plt.show()
```

Due to the "salt-and-pepper"-like distribution of the rice in the image, I would expect that the box filter would perform slight better at the blurring task, and therefore produce a sharper Detail image. From the detail images in the second row above, this result is observed. The resulting box-sharpened image is therefore slightly sharper compared to the Gauss-sharpened image.

## Section 3 - Edge Detection
### 3.1 - Sobel Edge Detector

```python
# apply edge-detector
sobel_rice_x = cv2.Sobel(img_rice, cv2.CV_64F, 1, 0, ksize=5)
sobel_rice_y = cv2.Sobel(img_rice, cv2.CV_64F, 0, 1, ksize=5)
sobel_box_x = cv2.Sobel(rice_sharp_box, cv2.CV_64F, 1, 0, ksize=5)
sobel_box_y = cv2.Sobel(rice_sharp_box, cv2.CV_64F, 0, 1, ksize=5)
sobel_gauss_x = cv2.Sobel(rice_sharp_gauss, cv2.CV_64F, 1, 0, ksize=5)
sobel_gauss_y = cv2.Sobel(rice_sharp_gauss, cv2.CV_64F, 0, 1, ksize=5)

# display images
plt.subplot(121), plt.imshow(sobel_rice_x, cmap="gray")
plt.title("Original X-Detector"), plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(sobel_rice_y, cmap="gray")
plt.title("Original Y-Detector"), plt.xticks([]), plt.yticks([])
plt.show()
plt.subplot(121), plt.imshow(sobel_box_x, cmap="gray")
plt.title("Box Filter X-Detector"), plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(sobel_box_y, cmap="gray")
plt.title("Box Filter Y-Detector"), plt.xticks([]), plt.yticks([])
plt.show()
plt.subplot(121), plt.imshow(sobel_gauss_x, cmap="gray")
plt.title("Gauss Filter X-Detector"), plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(sobel_gauss_y, cmap="gray")
plt.title("Gauss Filter Y-Detector"), plt.xticks([]), plt.yticks([])
plt.show()
```

The sharpened images result in clearer edge-detected results. This also confirms that the box filter is better for sharpening at sharpening this image than the Gaussian filter, because the edge-detection result is improved with the box-sharpened image compared to the Gauss-sharpened image.

```python
# get magnitudes
sobel_rice_mag = cv2.magnitude(sobel_rice_x, sobel_rice_y)
sobel_box_mag = cv2.magnitude(sobel_box_x, sobel_box_y)
sobel_gauss_mag = cv2.magnitude(sobel_gauss_x, sobel_gauss_y)

# threshold values
thresh1 = 200
thresh2 = 800

# create output images
ret, sobel_rice_thresh_low = cv2.threshold(sobel_rice_mag, thresh1, 255, cv2.THRESH_BINARY)
ret, sobel_rice_thresh_high = cv2.threshold(sobel_rice_mag, thresh2, 255, cv2.THRESH_BINARY)
ret, sobel_box_thresh_low = cv2.threshold(sobel_box_mag, thresh1, 255, cv2.THRESH_BINARY)
ret, sobel_box_thresh_high = cv2.threshold(sobel_box_mag, thresh2, 255, cv2.THRESH_BINARY)
ret, sobel_gauss_thresh_low = cv2.threshold(sobel_gauss_mag, thresh1, 255, cv2.THRESH_BINARY)
ret, sobel_gauss_thresh_high = cv2.threshold(sobel_gauss_mag, thresh2, 255, cv2.THRESH_BINARY)

# output images
plt.subplot(121), plt.imshow(sobel_rice_thresh_low, cmap="gray")
plt.title("Original, T = " + str(thresh1)), plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(sobel_rice_thresh_high, cmap="gray")
plt.title("Original, T = " + str(thresh2)), plt.xticks([]), plt.yticks([])
plt.show()
plt.subplot(121), plt.imshow(sobel_box_thresh_low, cmap="gray")
plt.title("Box Filtered, T = " + str(thresh1)), plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(sobel_box_thresh_high, cmap="gray")
plt.title("Box Filtered, T = " + str(thresh2)), plt.xticks([]), plt.yticks([])
plt.show()
plt.subplot(121), plt.imshow(sobel_gauss_thresh_low, cmap="gray")
plt.title("Gauss Filtered, T = " + str(thresh1)), plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(sobel_gauss_thresh_high, cmap="gray")
plt.title("Gauss Filtered, T = " + str(thresh2)), plt.xticks([]), plt.yticks([])
plt.show()
```

The low threshold results in an image that isn't very clear, because many of the points in between the grains of rice that still have some intensity are above the threshold. The high threshold results in an image that quite clearly outlines the grains of rice, because only the grains themselves are outlined. The box- and Gauss-sharpened images actually have a bit of a negative effect on these magnitude images. The original magnitude image is clearer than the filtered magnitude images, at both thresholds.
